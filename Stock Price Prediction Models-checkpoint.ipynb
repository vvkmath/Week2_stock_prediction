```python
#Regression modeling to forecast stock price for Google

#Import dependencies 
import pandas as pd  
import numpy as np  
import matplotlib.pyplot as plt  
import seaborn as seabornInstance 
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn import metrics
 
#Read the csv file with stock price history
dataset = pd.read_csv('C:/Siraj_Rival_ML_Course/ML_Course/Week2/Stock_price_prediction_homework/GOOG.csv')

#Explore dataset
dataset.shape
dataset.describe()
```

```python
#Add a time series ('Num_days') to track the number of days of trading.
import datetime
from datetime import date
dataset['Num_days'] = (pd.to_datetime(dataset['Date']) - pd.to_datetime(dataset['Date'][0])).dt.days.astype('int16') 
dataset.set_index('Num_days', inplace=True)
```

```python
#Compute the 100-day moving average to use as a feature in the regression model.
close_px = dataset['Adj Close']
mavg = close_px.rolling(window=100).mean()
dataset['MA_adj_close'] =mavg
dataset= dataset[dataset['MA_adj_close'].notnull()]
dataset.tail()
```

```python
#Plot the stock price over time   
dataset.plot(y='Adj Close', style='b', figsize=(15,6))  
plt.title('Adjusted Closing Price')  
plt.xlabel('Days')  
plt.ylabel('Stock Price')  
plt.show() 
```

```python
#Preprocess data for model inputs
indices = np.arange(len(dataset)).astype(np.int)[:,None]
y_data = dataset['Adj Close'].values[indices]  
x_data = dataset['MA_adj_close'].values[indices]  
X = x_data[:,]  
y = y_data[:,]  

#Create training and test datasets
split_pct = 0.8
split_ind = int(split_pct * len(dataset))
X_train = X[:split_ind]
y_train = y[:split_ind]
X_test = X[split_ind:]
y_test = y[split_ind:]
```

```python
#Model1: Quadratic Regression 3
from sklearn.linear_model import Ridge
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

clfpoly2 = make_pipeline(PolynomialFeatures(3), Ridge())
clfpoly2.fit(X_train, y_train)

y_pred = clfpoly2.predict(X_test)
df = pd.DataFrame({'Actual': y_test.flatten(), 'Predicted': y_pred.flatten()})
df= df.sort_index(ascending='false')
```

```python
#Plot the adj closing and predicted stock price over time 
fig, ax = plt.subplots(figsize=(15,7))
plt.plot('Actual', data=df, marker='o', markerfacecolor='blue', markersize=3, color='skyblue', linewidth=2)
plt.plot('Predicted', data=df, marker='', color='olive', linewidth=2)
plt.legend()
plt.title('Quadratic Regression Model: Date. vs Stock Price')  
plt.xlabel('Days')  
plt.ylabel('Stock Price')  
plt.show()
```

```python
#Model2: Gradient Boosting 
from sklearn.ensemble import GradientBoostingRegressor

gbt = GradientBoostingRegressor()
gbt.fit(X_train, y_train)

y_pred = gbt.predict(X_test)
df = pd.DataFrame({'Actual': y_test.flatten(), 'Predicted': y_pred.flatten()})
df= df.sort_index(ascending='false')
```

```python
#Plot the adj closing and predicted stock price over time 
fig, ax = plt.subplots(figsize=(15,7))
plt.plot('Actual', data=df, marker='o', markerfacecolor='blue', markersize=3, color='skyblue', linewidth=2)
plt.plot('Predicted', data=df, marker='', color='olive', linewidth=2)
plt.legend()
plt.title('Gradient Boosting Model: Date. vs Stock Price')  
plt.xlabel('Days')  
plt.ylabel('Stock Price')  
plt.show()
```

```python
#Model3: Gradient Boosting with past stock price data
#Use previous 1 month (30 days) stock prices as features to predict Adj.Closing price
time_period = 30 
indices = np.arange((len(dataset) - time_period)).astype(np.int)[:,None] + np.arange(time_period + 1).astype(np.int)

#Create training and test datasets for 30 days data
data = dataset['Adj Close'].values[indices]
X = data[:,:-1]  
y = data[:,-1]  
split_pct = 0.8

split_ind = int(split_pct * len(dataset))
X_train = X[:split_ind]
y_train = y[:split_ind]
X_test = X[split_ind:]
y_test = y[split_ind:]

from sklearn.ensemble import GradientBoostingRegressor

gbt = GradientBoostingRegressor()
gbt.fit(X_train, y_train)

y_pred = gbt.predict(X_test)
df = pd.DataFrame({'Actual': y_test.flatten(), 'Predicted': y_pred.flatten()})
df= df.sort_index(ascending='false')
```

```python
#Plot the adj closing and predicted stock price over time 
fig, ax = plt.subplots(figsize=(15,7))

plt.plot('Actual', data=df, marker='o', markerfacecolor='blue', markersize=3, color='skyblue', linewidth=2)
plt.plot('Predicted', data=df, marker='', color='olive', linewidth=2)
plt.legend()
plt.title('Gradient Boosting with 30-days data: Date. vs Stock Price')  
plt.xlabel('Days')  
plt.ylabel('Stock Price')  
plt.show()
```
